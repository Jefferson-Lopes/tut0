## Accumulation

Accumulation(Integration) is a sample way to redunction the very weak signals from noise, and which is the most widely used method in Astronomy.The noise integrates down as $\sqrt{t}$, whereas the signal is coherent.

In the following example, a very weak signal is submerged in a series of noise, neither human nor machine could recognise it. But, after a long time integration you can see a very strong pulse shown out. That's the magic of integration.

```python
%matplotlib inline
```

```python
# Import required modules
import numpy as np
import matplotlib.pyplot as plt
```

Most of the integration calculation happened at the end of signal processing, for example after FFT or correlation. Suppose we have 1024 FFT channels and you injected a single pulse signal in one of these channels, but this single pulse isn't strong enough to see(the first figure shows before integration, there is nothing in those channels). However, after integration the signal shows up(see the second figure of following).

```python
fft_p     = 1024       # Number of fft channels
n_int = 10000       # Number of integration times

samples = np.zeros(fft_p)
samples[320]=0.02  # Set a signal in one frequency channel

noise   = np.random.random(n_int*fft_p) # generate a noise signal

signal = n_int*(samples).tolist()
data = noise + signal # mix the signal with noise

data_int = np.array(data).reshape(n_int,fft_p)
data_int = data_int.T.mean(axis=1)

plt.subplot(2,1,1)
plt.title("Before Integration")
plt.plot(data[:1024],label='Signal + Noise') # plot signal before integration

plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=2.0)
plt.legend()
plt.subplot(2,1,2)
plt.title("After Integration")
plt.plot(data_int[:1024],label='Signal + Noise') # plot signal after integration

plt.legend()
```
