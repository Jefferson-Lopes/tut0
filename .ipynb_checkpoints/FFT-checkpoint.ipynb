## Fourier Tranformation(FFT)

The Fourier Transform (FT), in one form or another, is perhaps the most important single concept in astronomical instrumentation. It expresses the relation between a function and its spectrum, or between the autocorrelation of that function and its power spectrum. In astronomical instrumentation, the "function" is always the electromagnetic (e.m.) field (or an equivalent representation of it, such as a current, voltage, modulation of a carrier signal, etc.) . The independent variable of the function may be time, in which case the "spectrum" is indeed the kind of spectrum that telephone engineers and  recording technicians might talk about. Its discrete counterpart, the Discrete Fourier Transform (DFT), which is normally computed using the so-called Fast Fourier Transform (FFT), has revolutionized modern society, as it is ubiquitous in digital electronics and signal processing. 

### Here is an example for 1 Dimension Fourier Transformation

```python
%matplotlib inline
```

```python
import numpy as np
import numpy.fft as fft
import matplotlib as mpl

import matplotlib.pyplot as plt
mpl.rcParams['figure.figsize'] = (12, 5)
mpl.rcParams['image.cmap'] = "gray"
```

Generate a sine wave signal

```python
N = 4096  # number of samples
x = np.linspace(-np.pi/2, np.pi/2, N)
```

```python
# Define amplitude, frequency, and phase of a 1-D sine wave
A = 10
f = 8
phi = 30
y = A * np.sin(f * x + phi)
plt.plot(x, y); plt.title('Simple Sine Wave');
```

Plotting Real & Imaginary Parts of 1D FFT

```python
Y = fft.fft(y)
plt.subplot(1,2,1)
plt.plot(fft.fftfreq(N), Y.real); plt.xlim(-np.pi/32, np.pi/32); plt.title('Real');
plt.subplot(1,2,2)
plt.plot(fft.fftfreq(N), Y.imag); plt.xlim(-np.pi/32, np.pi/32); plt.title('Imag');
```

FFT "Standard" Order

```python
Y = fft.fft(y)
plt.subplot(1,2,1)
plt.plot(np.abs(Y))
plt.xlim(0,N)
plt.title('Raw FFT Output')
plt.subplot(1,2,2)
#plt.figure()
plt.plot(np.abs(fft.fftshift(Y)))
plt.xlim(0,N); plt.title('Shifted FFT Output')
```

Plotting 1D FFT Magnitude

```python
plt.plot(fft.fftfreq(N), np.abs(Y))
plt.title('Magnitude')
```

Mixed Sinusoid

```python
# Define amplitude, frequency, and phase of a 1-D cosine wave
A = [127, 63, 31]
F = [8, 64, 256]
Phi = [0, np.pi/8, np.pi/6]
y = np.sum([a * np.cos(f * x + phi) for a, f, phi in zip(A, F, Phi)], axis=0)
plt.plot(x, y)
```

Plotting Real & Imaginary Parts of 1D FFT

```python
Y = fft.fft(y)
plt.subplot(1,2,1)
plt.plot(fft.fftfreq(N), Y.real); plt.xlim(-np.pi/32, np.pi/32)
plt.subplot(1,2,2)
plt.plot(fft.fftfreq(N), Y.imag); plt.xlim(-np.pi/32, np.pi/32)
```
